# **Data Cleaning and Preprocessing**

*** *The dataset consists of 13 columns. The dataset has both Qualitative data and Quantitative data. Four out of 13 columns - Make, Model, Type, Origin, DriveTrain are nominal data which is of qualitative type (i.e) we cannot perform any arithmetic operations on them but can use them as label or classifiers. So we can perform mean on the other columns to fill the Null data.* ***

Since there majorly 4 null datas, we can take the mean of the total column and put it to them. It will not be a big difference as 4 is negligible compared to 428 rows of data. This operation is shown below 

```python
data[['MSRP','Invoice','EngineSize','Cylinders','Horsepower','MPG_City','MPG_Highway','Weight','Wheelbase','Length']] = data[['MSRP','Invoice','EngineSize','Cylinders','Horsepower','MPG_City','MPG_Highway','Weight','Wheelbase','Length']].fillna(data[['MSRP','Invoice','EngineSize','Cylinders','Horsepower','MPG_City','MPG_Highway','Weight','Wheelbase','Length']].mean())
```

```python
data.isnull().sum()
```

From the above Null data count output we can see that the MSRP and Invoice columns still contains the null values even after replacing the mean values. By checking the dataset once again, those two columnns contains $ symbol which makes them a "String data type"(Something Like a word). So we need to remove the symble. The code for removing the symbol and filling null values with mean value is given below 

```python
data['MSRP'] = data['MSRP'].replace("[$,]", "", regex=True).astype(float)        #removes the $symbol in column MSRP
data['Invoice'] = data['Invoice'].replace("[$,]", "", regex=True).astype(float)  #removes the $symbol in column Invoice
data[['MSRP','Invoice']]= data[['MSRP','Invoice']].fillna(data[['MSRP','Invoice']].mean())
```

```python
data.isnull().sum()
```

From the above output, the MSRP and Invoice columns are sorted. Now we need to work on the nominal data. Since 4 datas are negligible compared to 428, we fill the 4 null data with most occured data. This method works well when there is a negligible null data( like we have in our case) 

```python
clean_data = data.apply(lambda x: x.fillna(x.value_counts().index[0]))
clean_data
```

Checking for null value


```python
clean_data.isnull().sum()
```

Analyzing different Car Makes

```python
clean_data['Make'].value_counts()
```

Looking for cars from Asia and Europe

```python
origin_cars = clean_data[clean_data['Origin'].isin(['Asia','Europe'])] #this checks for the cars that are in asia and europe
origin_cars['Make'].value_counts()
```

```python
origin_cars
```

Filtering cars which has weight lesser than 4000 from the region of Asia and Europe

```python
filtered_cars=origin_cars[~(origin_cars['Weight'] > 4000)] #this filters the cars with weight less than 4000
filtered_cars
```
